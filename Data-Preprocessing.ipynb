```python
import os
import cv2


DATASET_PATH = "D:\OCR\docs-sm"

# Select only these categories
selected_categories = ["invoice", "email", "budget"]

valid_files = []
corrupt_files = []

# Loop through only selected folders
for category in selected_categories:
    category_path = os.path.join(DATASET_PATH, category)
    
    if os.path.isdir(category_path):  # Ensure it's a directory
        for file in os.listdir(category_path):
            file_path = os.path.join(category_path, file)

            # Process only image files
            if file.lower().endswith((".jpg", ".jpeg", ".png", ".tiff", ".bmp", ".gif", ".webp")):
                img = cv2.imread(file_path)
                if img is not None:
                    valid_files.append({"file_path": file_path, "label": category})
                else:
                    corrupt_files.append(file_path)

print(f"Valid files count: {len(valid_files)}")
print(f"Corrupt files count: {len(corrupt_files)}")

# Remove corrupted files
for file in corrupt_files:
    os.remove(file)
    print(f"Removed corrupted file: {file}")
```

```python
import pandas as pd

df = pd.DataFrame(valid_files)

# Save to CSV for reference
df.to_csv("filtered_dataset.csv", index=False)
print("✅ Filtered dataset saved as filtered_dataset.csv")
```

```python
import pytesseract
import pandas as pd

# If using Windows, set Tesseract path (modify this according to your installation)
pytesseract.pytesseract.tesseract_cmd = r"C:\Program Files\Tesseract-OCR\tesseract.exe"

# Function to extract text from an image
def extract_text_tesseract(image_path):
    img = cv2.imread(image_path)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)  # Convert to grayscale
    text = pytesseract.image_to_string(gray)  # Perform OCR
    return text.strip()

# Process all filtered images
extracted_data = []

for index, row in df.iterrows():  
    image_path = row["file_path"]
    label = row["label"]
    
    extracted_text = extract_text_tesseract(image_path)
    extracted_data.append({"file_path": image_path, "label": label, "text": extracted_text})

# Convert to DataFrame
df_extracted = pd.DataFrame(extracted_data)

# Save extracted text
df_extracted.to_csv("extracted_texts_tesseract.csv", index=False)
print("✅ Text extraction completed! Saved to extracted_texts_tesseract.csv")
```

```python
def classify_text(model, text):
    prompt = f"""
    Classify the following document into one of these categories: 'invoice', 'email', or 'budget'.
    
    Document text:
    {text[:1000]}

    Output only the category name without any explanation.
    """

    response = model(prompt, max_new_tokens=10)
    return response.strip()
```

```python
import pandas as pd
import re
import nltk
from nltk.corpus import stopwords
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report

# Download stopwords
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

# Load OCR dataset
df = pd.read_csv("extracted_texts_tesseract.csv")  # Ensure file exists

# Fill missing values
df["text"] = df["text"].fillna("")

# Text cleaning function
def preprocess_text(text):
    text = text.lower()  # Convert to lowercase
    text = re.sub(r'\d+', '', text)  # Remove numbers
    text = re.sub(r'\W+', ' ', text)  # Remove special characters
    text = ' '.join([word for word in text.split() if word not in stop_words])  # Remove stopwords
    return text

# Apply text preprocessing
df["clean_text"] = df["text"].apply(preprocess_text)

print("✅ Text Preprocessing Done!")
```

```python
# Split dataset into training and testing
X_train, X_test, y_train, y_test = train_test_split(df["clean_text"], df["label"], test_size=0.2, random_state=42)

# Convert text to numerical features using TF-IDF (with N-grams)
vectorizer = TfidfVectorizer(max_features=5000, ngram_range=(1, 2))  # Includes unigrams and bigrams
X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)

# Train Naïve Bayes model
model = MultinomialNB(alpha=0.1)  # Alpha = 0.1 works well for text classification
model.fit(X_train_tfidf, y_train)

# Make predictions
y_pred = model.predict(X_test_tfidf)

# Check accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"✅ Naïve Bayes Model Trained! Accuracy: {accuracy * 100:.2f}%")

# Print detailed classification report
print(classification_report(y_test, y_pred))
```

```python
import pickle
import pandas as pd
import re
import nltk
from nltk.corpus import stopwords

# Load trained model
with open("naive_bayes_model.pkl", "rb") as model_file:
    model = pickle.load(model_file)

# Load TF-IDF vectorizer
with open("tfidf_vectorizer.pkl", "rb") as vectorizer_file:
    vectorizer = pickle.load(vectorizer_file)

# Download stopwords
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

# Load your existing dataset
df = pd.read_csv("extracted_texts_tesseract.csv")  # Use the same dataset for classification

# Fill missing values
df["text"] = df["text"].fillna("")

# Text preprocessing function
def preprocess_text(text):
    text = text.lower()
    text = re.sub(r'\d+', '', text)  # Remove numbers
    text = re.sub(r'\W+', ' ', text)  # Remove special characters
    text = ' '.join([word for word in text.split() if word not in stop_words])  # Remove stopwords
    return text

# Apply text preprocessing
df["clean_text"] = df["text"].apply(preprocess_text)

# Convert text to numerical features using the trained vectorizer
X_tfidf = vectorizer.transform(df["clean_text"])

# Predict document types
df["predicted_label"] = model.predict(X_tfidf)

# Save classified documents
df.to_csv("classified_documents.csv", index=False)

print("✅ Classification completed! Check 'classified_documents.csv' for results.")
```
